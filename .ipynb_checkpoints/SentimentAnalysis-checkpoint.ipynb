```python
import os
import pandas as pd
import re
import emoji
from collections import Counter

df_file = os.path.join("datasets", "train-balanced-sarcasm.csv")

# Funzione per pulire il testo
def clean_text(text):
    if pd.isna(text):
        return ""
    text = str(text)
    text = re.sub(r'http\S+', '', text)  # Rimuove URL
    text = re.sub(r'@\w+', '', text)  # Rimuove menzioni
    text = re.sub(r'#\w+', '', text)  # Rimuove hashtag
    text = re.sub(r'\d+', '', text)  # Rimuove numeri
    text = re.sub(r'[^\w\s]', '', text)  # Rimuove punteggiatura
    return text.lower().strip()

# Funzione per verificare se il testo è valido
def is_valid_text(text):
    # Rimuove emoji
    text = emoji.replace_emoji(text, '')
    # Rimuove spazi extra
    text = re.sub(r'\s+', ' ', text).strip()
    # Verifica se il testo ha almeno 3 parole e più di 10 caratteri
    words = text.split()
    
    # Verifica se ci sono parole ripetute più di 5 volte
    word_counts = Counter(words)
    if any(count > 5 for count in word_counts.values()):
        return False
    
    return len(words) >= 3 and len(text) > 10 and not text.isnumeric()

# Carica i dati
print("Caricamento dei dati...")
df = pd.read_csv(df_file)
df = df[["label", "comment"]]

# Pulizia del dataset
print("Pulizia del dataset in corso...")
df['cleaned_comment'] = df['comment'].apply(clean_text)
df['is_valid'] = df['cleaned_comment'].apply(is_valid_text)
df = df[df['is_valid']]
df = df.drop('is_valid', axis=1)
df['comment'] = df['comment'].astype(str)

# Campionamento
sample = 200000
df = df.head(sample)

df.to_csv("train-balanced-sarcasm-cleaned.csv", index=False)

print("Dataset pulito e filtrato. Dimensioni del nuovo DataFrame:", df.shape)
print("\nPrime 5 righe del DataFrame pulito:")
print(df.head())
```

```python
import pandas as pd
from tqdm import tqdm

# Function to truncate text to the maximum length accepted by the model
def truncate_text(text, max_length=512):
    words = text.split()
    if len(words) > max_length:
        return " ".join(words[:max_length])
    return text

# Function to analyze sentiment considering sarcasm
def analyze_sentiment_with_sarcasm(text, is_sarcastic):
    try:
        truncated_text = truncate_text(text)
        result = sentiment_pipeline(truncated_text)[0]
        label = result['label']
        
        if is_sarcastic:
            # Invert sentiment for sarcastic comments
            if label == 'LABEL_0':
                return 'Positive'
            elif label == 'LABEL_2':
                return 'Negative'
            else:
                return 'Neutral'
        else:
            if label == 'LABEL_0':
                return 'Negative'
            elif label == 'LABEL_1':
                return 'Neutral'
            else:
                return 'Positive'
    except Exception as e:
        print(f"Error in sentiment analysis for text: {text[:50]}...")
        print(f"Error: {str(e)}")
        return "Error"

# Function to process the dataframe
def process_dataframe(df):
    tqdm.pandas()
    
    # Analyze sentiment considering sarcasm
    df['sentiment'] = df.progress_apply(lambda row: analyze_sentiment_with_sarcasm(row['cleaned_comment'], row['label'] == 1), axis=1)
    
    return df

# Process the dataframe
print("Analyzing sentiment...")
df_processed = process_dataframe(df)

# Show sentiment distribution
print("\nSentiment distribution:")
print(df_processed['sentiment'].value_counts())

# Show some examples
print("\nSome analysis examples:")
sample_size = min(10, len(df_processed))
for _, row in df_processed.sample(sample_size).iterrows():
    print(f"Original comment: {row['comment'][:100]}...")
    print(f"Cleaned comment: {row['cleaned_comment'][:100]}...")
    print(f"Sentiment: {row['sentiment']}")
    print(f"Sarcastic: {'Yes' if row['label'] == 1 else 'No'}\n")

# Save the result
output_file = 'df_with_sentiment_and_sarcasm.csv'
df_processed.to_csv(output_file, index=False)
print(f"\nDataset with sentiment analysis considering sarcasm saved in {output_file}")

# Final statistics
print(f"\nTotal rows: {len(df_processed)}")
print(f"Sentiment distribution:")
print(df_processed['sentiment'].value_counts(normalize=True).mul(100).round(2).astype(str) + '%')
```

```python
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')

def get_sentiment_scores(text):
    sia = SentimentIntensityAnalyzer()
    return sia.polarity_scores(text)

def analyze_sentiment_with_sarcasm(row):
    comment = row['comment']  # Assumo che la colonna si chiami 'cleaned_comment'
    is_sarcastic = row['label'] == 1  # Assumo che la colonna del sarcasmo si chiami 'label'
    base_sentiment = row['sentiment']  # Assumo che la colonna del sentiment si chiami 'sentiment'
    
    sentiment_scores = get_sentiment_scores(comment)
    compound_score = sentiment_scores['compound']
    
    if is_sarcastic:
        if base_sentiment == 'positive':
            return 'Negative' if compound_score < 0.2 else 'Positive'
        elif base_sentiment == 'negative':
            return 'Positive' if compound_score > -0.2 else 'Negative'
        else:  # neutral
            return 'Negative' if compound_score < 0 else 'Positive'
    else:
        return 'Positive' if compound_score >= 0 else 'Negative'

def process_dataframe(df):
    df['Refined_Sentiment'] = df.apply(analyze_sentiment_with_sarcasm, axis=1)
    return df

# Carica il dataset
print("Caricamento del dataset...")
df = pd.read_csv(os.path.join("datasets", "df_with_sentiment.csv"))

# Verifica le colonne presenti nel dataset
print("\nColonne presenti nel dataset:")
print(df.columns)

# Processa il dataframe
print("\nProcessing del dataframe in corso...")
df_processed = process_dataframe(df)

# Mostra alcuni risultati
print("\nPrimi 10 risultati dell'analisi:")
print(df_processed[['cleaned_comment', 'label', 'sentiment', 'Refined_Sentiment']].head(10))

# Statistiche finali
print("\nDistribuzione del Refined Sentiment:")
print(df_processed['Refined_Sentiment'].value_counts(normalize=True).mul(100).round(2).astype(str) + '%')

# Salva i risultati
output_file = 'df_with_refined_sentiment.csv'
df_processed.to_csv(output_file, index=False)
print(f"\nDataset con analisi del sentiment raffinata salvato in {output_file}")

# Statistiche aggiuntive
print("\nConfronti tra Sentiment originale e Refined Sentiment:")
comparison = pd.crosstab(df_processed['sentiment'], df_processed['Refined_Sentiment'], normalize='index')
print(comparison.round(2))

print("\nPercentuale di cambiamenti nel sentiment:")
changes = (df_processed['sentiment'] != df_processed['Refined_Sentiment']).mean() * 100
print(f"{changes:.2f}% dei commenti ha cambiato sentiment dopo l'analisi raffinata")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
import os

def load_data(file_path):
    print("Loading CSV file...")
    return pd.read_csv(file_path)

def analyze_sentiment_distribution(df, column, title):
    dist = df[column].value_counts(normalize=True).mul(100).round(2)
    print(f"\n{title} Distribution:")
    print(dist.astype(str) + '%')
    return dist

def plot_sentiment_comparison(sarcastic_dist, non_sarcastic_dist, title):
    fig, ax = plt.subplots(figsize=(10, 6))
    width = 0.35
    x = range(len(sarcastic_dist))
    
    ax.bar([i - width/2 for i in x], sarcastic_dist.values, width, label='Sarcasm', color='lightcoral')
    ax.bar([i + width/2 for i in x], non_sarcastic_dist.values, width, label='Non-Sarcasm', color='lightskyblue')
    
    ax.set_ylabel('Percentage')
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(sarcastic_dist.index)
    ax.legend()
    
    plt.tight_layout()
    os.makedirs('graphs', exist_ok=True)
    plt.savefig(f"graphs/{title.replace(' ', '_')}.png")
    print(f"Graph saved as graphs/{title.replace(' ', '_')}.png")

df = load_data('df_with_sentiment.csv')

# Sentiment analysis ignoring sarcasm
sentiment_dist = analyze_sentiment_distribution(df, 'sentiment', 'Sentiment (Ignoring Sarcasm)')

# Sentiment analysis considering sarcasm
df['adjusted_sentiment'] = df.apply(lambda row: 
    'Negative' if row['sentiment'] == 'Positive' and row['label'] == 1 else
    'Positive' if row['sentiment'] == 'Negative' and row['label'] == 1 else
    row['sentiment'], axis=1)
adjusted_sentiment_dist = analyze_sentiment_distribution(df, 'adjusted_sentiment', 'Sentiment (Considering Sarcasm)')

# Comparison between sarcastic and non-sarcastic comments
sarcastic_df = df[df['label'] == 1]
non_sarcastic_df = df[df['label'] == 0]

sarcastic_dist = analyze_sentiment_distribution(sarcastic_df, 'adjusted_sentiment', 'Sentiment of Sarcastic Comments')
non_sarcastic_dist = analyze_sentiment_distribution(non_sarcastic_df, 'sentiment', 'Sentiment of Non-Sarcastic Comments')

# Creating graphs
plot_sentiment_comparison(sentiment_dist, adjusted_sentiment_dist, 'Sentiment Comparison: Ignoring vs Considering Sarcasm')
plot_sentiment_comparison(sarcastic_dist, non_sarcastic_dist, 'Sentiment Comparison: Sarcastic vs Non-Sarcastic Comments')

# Analyzing the impact of sarcasm
sarcasm_impact = df[df['sentiment'] != df['adjusted_sentiment']]
print(f"\nNumber of comments where sarcasm changed the sentiment: {len(sarcasm_impact)} ({len(sarcasm_impact)/len(df)*100:.2f}%)")

# Examples of comments where sarcasm changed the sentiment
print("\nExamples of comments where sarcasm changed the sentiment:")
for _, row in sarcasm_impact.sample(min(5, len(sarcasm_impact))).iterrows():
    print(f"Comment: {row['cleaned_comment'][:100]}...")
    print(f"Original sentiment: {row['sentiment']}")
    print(f"Adjusted sentiment: {row['adjusted_sentiment']}\n")

# Saving the results
df.to_csv('df_with_comparative_sentiment.csv', index=False)
print("Dataset with comparative sentiment analysis saved in df_with_comparative_sentiment.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
import os

def load_data(file_path):
    print(f"Loading file {file_path}...")
    return pd.read_csv(file_path)

def analyze_sentiment_distribution(df, column, title):
    dist = df[column].value_counts(normalize=True).mul(100).round(2)
    print(f"\nDistribution of {title}:")
    print(dist.astype(str) + '%')
    return dist

def plot_sentiment_comparison(dist1, dist2, title, label1, label2):
    fig, ax = plt.subplots(figsize=(10, 6))
    width = 0.35
    x = range(len(dist1))
    
    ax.bar([i - width/2 for i in x], dist1.values, width, label=label1, color='lightcoral')
    ax.bar([i + width/2 for i in x], dist2.values, width, label=label2, color='lightskyblue')
    
    ax.set_ylabel('Percentage')
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(dist1.index)
    ax.legend()
    
    plt.tight_layout()
    
    # Create the graphs/comparison directory if it doesn't exist
    os.makedirs('graphs/comparison', exist_ok=True)
    
    file_path = f"graphs/comparison/{title.replace(' ', '_')}.png"
    plt.savefig(file_path)
    print(f"Graph saved as {file_path}")

# Load both datasets
df_original = pd.read_csv(os.path.join("datasets", "df_with_sentiment.csv"))
df_adjusted = pd.read_csv(os.path.join("datasets", "df_with_adjusted_sentiment.csv"))

# Ensure the dataframes are aligned
assert len(df_original) == len(df_adjusted), "The datasets do not have the same number of rows"

# Analysis of original and adjusted sentiment
original_sentiment_dist = analyze_sentiment_distribution(df_original, 'sentiment', 'Original Sentiment (Ignoring Sarcasm)')
adjusted_sentiment_dist = analyze_sentiment_distribution(df_adjusted, 'sentiment', 'Adjusted Sentiment (Considering Sarcasm)')

# Comparison between original and adjusted sentiment
plot_sentiment_comparison(original_sentiment_dist, adjusted_sentiment_dist, 
                          'Sentiment Comparison: Ignoring vs Considering Sarcasm', 
                          'Ignoring Sarcasm', 'Considering Sarcasm')

# Comparison between sarcastic and non-sarcastic comments
sarcastic_df = df_adjusted[df_adjusted['label'] == 1]
non_sarcastic_df = df_adjusted[df_adjusted['label'] == 0]

sarcastic_dist = analyze_sentiment_distribution(sarcastic_df, 'sentiment', 'Sentiment of Sarcastic Comments (Adjusted)')
non_sarcastic_dist = analyze_sentiment_distribution(non_sarcastic_df, 'sentiment', 'Sentiment of Non-Sarcastic Comments')

plot_sentiment_comparison(sarcastic_dist, non_sarcastic_dist, 
                          'Sentiment Comparison: Sarcastic vs Non-Sarcastic Comments',
                          'Sarcastic Comments', 'Non-Sarcastic Comments')

# Analysis of sarcasm impact
sarcasm_impact = df_original[df_original['sentiment'] != df_adjusted['sentiment']]
print(f"\nNumber of comments where sarcasm changed the sentiment: {len(sarcasm_impact)} ({len(sarcasm_impact)/len(df_original)*100:.2f}%)")

# Examples of comments where sarcasm changed the sentiment
print("\nExamples of comments where sarcasm changed the sentiment:")
for idx, row in sarcasm_impact.sample(min(5, len(sarcasm_impact))).iterrows():
    print(f"Comment: {row['cleaned_comment'][:100]}...")
    print(f"Original sentiment: {row['sentiment']}")
    print(f"Adjusted sentiment: {df_adjusted.loc[idx, 'sentiment']}")
    print(f"Sarcastic: Yes\n")

# Additional analysis: sarcasm distribution
sarcasm_dist = df_original['label'].value_counts(normalize=True).mul(100).round(2)
print("\nSarcasm Distribution:")
print(f"Non-Sarcastic: {sarcasm_dist[0]:.2f}%")
print(f"Sarcastic: {sarcasm_dist[1]:.2f}%")

# Creation of a combined dataframe for final analysis
df_combined = df_original.copy()
df_combined['adjusted_sentiment'] = df_adjusted['sentiment']
df_combined.to_csv('df_with_comparative_sentiment.csv', index=False)
print("\nCombined dataset with original and adjusted sentiment saved in df_with_comparative_sentiment.csv")
```
